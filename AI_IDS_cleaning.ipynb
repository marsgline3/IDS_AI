```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('datasets/Wednesday.csv')
```

```python
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.dropna(inplace=True)
data.info()
```

```python
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
data.info()
data[' Label'].value_counts()
```

```python
mask = data[' Label'] == 'Heartbleed'
htbleed= data[mask]
data.drop(htbleed.index, inplace=True)
data[' Label'].value_counts()
```

```python
data[' Label'] = data[' Label'].replace({'BENIGN': 0})
data[' Label'] = data[' Label'].where(data[' Label'] == 0, 1)
data[' Label'].describe()
```

```python
scaler = MinMaxScaler()
scaler.fit_transform(data)
data.describe()
```

```python
data[' Label'].value_counts()
```

```python
data.to_csv('datasets/Wednesday_cleaned.csv', index=False)
```
